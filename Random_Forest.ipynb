```python
from sklearn.ensemble import RandomForestRegressor
import tensorflow as tf
```

# ชุด1

```python
data_1.drop('Unnamed: 0',1,inplace = True)
data_1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Scaler1 = scaler.fit_transform(data_1)
```

```python
data_1 = pd.DataFrame(Scaler1)
data_1 = data_1.rename(columns={0: 'Sex' ,1: 'AgeSYear',2:'telomere length', 3:'Weight',4:'Height',5:'BMI',6:'Systolic',7:'Diastolic',8:'HeartRate',9:'SMM',10:'Fat Mass',
                                11:'% Body fat',12:'waist to hip radio',13:'abdominal circumference',14:'visceral fat area',15:'Glucose',16:'BUN',17:'Creatinine',18:'Uric Acid',19:'Cholesterol',
                                20:'Triglyceride',21:'HDL-C',22:'LDL',23:'AST',24:'ALT',25:'Alkaline Phos',26:'HbA1c'})
```

```python
data_1
```

```python
y1 = data_1[['telomere length']]
X1 = data_1.drop('telomere length',1)
```

```python
data_1 = pd.DataFrame(Scaler1)
data_1 = data_1.rename(columns={0: 'Sex' ,1: 'AgeSYear',2:'telomere length', 3:'Weight',4:'Height',5:'BMI',6:'Systolic',7:'Diastolic',8:'HeartRate',9:'SMM',10:'Fat Mass',
                                11:'% Body fat',12:'waist to hip radio',13:'abdominal circumference',14:'visceral fat area',15:'Glucose',16:'BUN',17:'Creatinine',18:'Uric Acid',19:'Cholesterol',
                                20:'Triglyceride',21:'HDL-C',22:'LDL',23:'AST',24:'ALT',25:'Alkaline Phos',26:'HbA1c'})
```

```python
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=99)
```

```python
tf.random.set_seed(99)
rg1 = RandomForestRegressor(n_estimators=50, max_depth = 3, max_leaf_nodes = 9, max_features = None, random_state=99)
rg1.fit(X1_train, y1_train)
```

```python
tf.random.set_seed(99)
y1_pred = rg1.predict(X1_test)
y1_pred= y1_pred.reshape(31,1)
```

```python
from sklearn import metrics
#Train
print('R2 Value Train:',metrics.r2_score(y1_train, rg1.predict(X1_train)))
print('MSE Train:',metrics.mean_squared_error(y1_train, rg1.predict(X1_train)))
print('MAE Train:',metrics.mean_absolute_error(y1_train, rg1.predict(X1_train)))
```

```python
import numpy as np
print('Accuracy', 100- (np.mean(np.abs((y1_test - y1_pred) / y1_test)) * 100))
```

```python
#Test
print('R2 Value Test:',metrics.r2_score(y1_test, y1_pred))
print('MSE Test:',metrics.mean_squared_error(y1_test, y1_pred))
print('MAE Test:',metrics.mean_absolute_error(y1_test, y1_pred))
```
